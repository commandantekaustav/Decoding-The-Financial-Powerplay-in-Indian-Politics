```python
# Sending a request to the website
url = "https://www.myneta.info/party/index.php?action=summary&id=9"
response = requests.get(url)

# Creating a BS object from the response html
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
parties = soup.find_all("div", class_="partyexpensebox")
for party in parties:
    party_name = party.find("div", class_="partyname").text.strip()
    expenses = party.find("span", class_="electionexpensetable").find_all("td")
    year_expenses = []
    for expense in expenses:
        year_expenses.append(expense.text.strip())
    print(party_name, year_expenses)
```
